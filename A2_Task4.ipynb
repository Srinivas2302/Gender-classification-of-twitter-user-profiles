```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('twitter_user_data.csv', encoding_errors = 'ignore')
```

```python
df.tail()
```

```python
df.info()
```

```python
#Replace unkown and null values from gender
df['gender'].replace({'unknown': None},inplace =True)
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
```

```python
df['gender_new'] = np.where(df['gender'] == 'female', 'Humans', np.where(df['gender'] == 'male', 'Humans', 'Non-Humans'))
```

```python
df.groupby('gender_new').agg(Tweet_avg=('tweet_count', 'mean'), retweet_avg=('retweet_count', 'mean'))

```

We can see non-Human accounts usually have higher number of tweets and re-tweets in comparison to Human accounts.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['sidebar_color'] = le.fit_transform(df['sidebar_color'])
```

```python
df.groupby('gender_new').agg(sidebar_color_avg=('sidebar_color', 'mean'))
```

Since color after label encoding is in similar range for both gender classes, color woluld not be a good decision variable for distinguishing thhem.
