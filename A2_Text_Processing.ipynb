## Replacing Unknown with null and after that replacing null with mode

```python
df['gender'].replace({'unknown': None},inplace =True)
```

```python
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
```

## As we have to distingusish between humans and Non Human, we will replace male and female with 0(Humans) and other category as 0(Non Human)

```python
df['gender_new'] = np.where(df['gender'] == 'female', 0, np.where(df['gender'] == 'male', 0, 1))
```

```python
#Importing libraries related to Text processing 

import re
import nltk as nlp
from nltk.corpus import stopwords
lemma = nlp.WordNetLemmatizer() 

nlp.download('punkt')
nlp.download('stopwords')
nlp.download('wordnet')
nlp.download('omw-1.4')
```

```python
#Text processing


text_cleaned = []            # empty list
for each in df['text']:
    each = re.sub("[^a-zA-Z]"," ", str(each))                                        # regex(regular expressions) for cleaning
    each = each.lower()                                                              # MAking sure everything is lowercase
    each = nlp.word_tokenize(each)                                                   # Use of tokenisation for split
    each = [word for word in each if not word in set(stopwords.words("english"))]    # Removal of stop words
    lemma = nlp.WordNetLemmatizer() 
    each = [lemma.lemmatize(i) for i in each]                                       # Words will be transformed to their roots
    each = " ".join(each)                                                           
    text_cleaned.append(each)
```

```python
#Defining max number of features as 600
max_features = 600

vectorizer = TfidfVectorizer(max_features=max_features,stop_words = "english")

sparce_matrix = vectorizer.fit_transform(text_cleaned).toarray()
words = vectorizer.get_feature_names()
```

```python
print("{} features used are {}".format(max_features,vectorizer.get_feature_names()))
```

```python
#Label Encoding Gender

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['gender_new'] = le.fit_transform(df['gender_new'])
```

```python
y = df.gender_new.values
x = sparce_matrix
```

```python
#Creating Training and Test Set

from sklearn.model_selection import train_test_split 

train_x, test_x, train_y, test_y = train_test_split(x,y,test_size=0.2)
```

```python
#Running Logistic Regression
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter = 2000)
lr.fit(train_x,train_y)
# prediction

y_pred = lr.predict(test_x)
accuracy = accuracy_score(test_y, y_pred)
print("Accuracy is: %.4f\n" % accuracy)
```
